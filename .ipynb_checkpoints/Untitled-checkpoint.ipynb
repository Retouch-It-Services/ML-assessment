```python
# -------------------------------------------------------------
# BLOCK 1 — IMPORTS & CONFIG
# -------------------------------------------------------------
import os
import numpy as np
import pandas as pd
from datetime import timedelta

from sklearn.experimental import enable_iterative_imputer  # Required before importing IterativeImputer
from sklearn.model_selection import train_test_split
from sklearn.impute import KNNImputer, IterativeImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.calibration import CalibratedClassifierCV, calibration_curve
from sklearn.metrics import brier_score_loss, roc_auc_score, precision_recall_fscore_support
from sklearn.feature_selection import mutual_info_classif
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
import shap

# Reproducibility
RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

# Helper: Net Benefit (Decision Curve)
def net_benefit(y_true, probs, threshold):
    preds = (probs >= threshold).astype(int)
    tp = ((preds == 1) & (y_true == 1)).sum()
    fp = ((preds == 1) & (y_true == 0)).sum()
    n = len(y_true)
    w = threshold / (1 - threshold)
    return (tp / n) - (w * fp / n)
```

```python
# -------------------------------------------------------------
# BLOCK 2 — LOAD & CLEAN DATA
# -------------------------------------------------------------
data_path = "diabetic_data.csv"  # Change path if needed

# Try reading safely
try:
    df = pd.read_csv(data_path, encoding="utf-8")
except UnicodeDecodeError:
    df = pd.read_csv(data_path, encoding="ISO-8859-1")
except pd.errors.ParserError:
    df = pd.read_csv(data_path, sep=",", on_bad_lines="skip", engine="python")

print("✅ Dataset loaded successfully")
print(f"Shape: {df.shape}")
print(df.head(3))

# Drop useless or high-missing columns
drop_cols = ['encounter_id', 'patient_nbr', 'weight', 'payer_code', 'medical_specialty']
df.drop(columns=[c for c in drop_cols if c in df.columns], inplace=True, errors='ignore')

# Replace '?' with NaN
df.replace("?", np.nan, inplace=True)

# Target: readmitted (<30 → 1, otherwise → 0)
df['readmitted'] = df['readmitted'].apply(lambda x: 1 if x in ['<30'] else 0)

print(f"Total patients: {df.shape[0]}, Readmitted: {df['readmitted'].sum()}")
```
