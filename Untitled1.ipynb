```python

import pandas as pd
import zipfile
import warnings

warnings.filterwarnings("ignore")

RANDOM_STATE = 42
TOP_K_FEATURES = 50
ZIP_PATH = r"C:\Users\Lenovo\Downloads\ai tasks\health\diabetes+130-us+hospitals+for+years+1999-2008.zip"
CSV_NAME_IN_ZIP = "diabetic_data.csv"

# Load CSV from ZIP
with zipfile.ZipFile(ZIP_PATH) as z:
    with z.open(CSV_NAME_IN_ZIP) as f:
        df = pd.read_csv(f)

# Quick check
print("Dataset shape:", df.shape)
print(df.head())

```

```python
# Target column
target = 'readmitted'  # this is what we want to predict

# Separate features and target
X = df.drop(columns=[target])
y = df[target]

# Quick inspection
print("Feature matrix shape:", X.shape)
print("Target vector shape:", y.shape)
print("\nUnique classes in target:", y.unique())
print("\nClass distribution:\n", y.value_counts())
```

```python
X.columns = [str(c).replace('[','_')
                     .replace(']','_')
                     .replace('<','_')
                     .replace('>','_')
                     .replace(' ','_') for c in X.columns]

print("Cleaned feature names:")
print(X.columns.tolist())
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode target variable to numeric
le = LabelEncoder()
y_enc = le.fit_transform(y)

print("\nEncoded target classes:")
for i, cls in enumerate(le.classes_):
    print(f"{cls} -> {i}")
```

```python

from sklearn.model_selection import TimeSeriesSplit
import numpy as np

# Optional: If the dataset has a time column, sort by it to simulate real-world scenario
# Example: df = df.sort_values('admission_date')

# Using 80/20 split based on index to avoid leakage
split_index = int(0.8 * len(X))
X_train, X_test = X.iloc[:split_index, :], X.iloc[split_index:, :]
y_train, y_test = y_enc[:split_index], y_enc[split_index:]

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer

# Identify numerical and categorical columns
num_cols = X_train.select_dtypes(include=['int64','float64']).columns.tolist()
cat_cols = X_train.select_dtypes(include=['object']).columns.tolist()

# Pipelines
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

# Combine pipelines
preprocessor = ColumnTransformer([
    ('num', num_pipeline, num_cols),
    ('cat', cat_pipeline, cat_cols)
])

# Fit-transform training data and transform test data
X_train_proc = preprocessor.fit_transform(X_train)
X_test_proc = preprocessor.transform(X_test)

print("Processed feature matrix shape:", X_train_proc.shape)
```

```python

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=200, 
    random_state=RANDOM_STATE, 
    n_jobs=-1
)
rf.fit(X_train_proc, y_train)
print("Random Forest model trained.")
```

```python

from xgboost import XGBClassifier

xgb_model = XGBClassifier(
    n_estimators=200,
    learning_rate=0.1,
    max_depth=5,
    random_state=RANDOM_STATE,
    use_label_encoder=False,
    eval_metric='mlogloss'
)
xgb_model.fit(X_train_proc, y_train)
print("XGBoost model trained.")
```

```python
# -------------------------------
# Multiclass ROC plots (one-vs-rest) for RF and XGBoost
# -------------------------------
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_auc_score, RocCurveDisplay, roc_curve, auc
import matplotlib.pyplot as plt
import numpy as np

# Ensure y_test is numeric encoded (y_test already encoded earlier)
classes = np.unique(y_test)
n_classes = len(classes)
print("Detected classes:", classes)

# Binarize the test labels for one-vs-rest ROC plotting
y_test_bin = label_binarize(y_test, classes=classes)  # shape (n_samples, n_classes)

def get_proba_or_decision(estimator, X):
    # prefer predict_proba, fallback to decision_function if needed
    if hasattr(estimator, "predict_proba"):
        return estimator.predict_proba(X)
    elif hasattr(estimator, "decision_function"):
        # decision_function might be shape (n_samples, n_classes) or (n_samples,)
        dec = estimator.decision_function(X)
        # if binary and returns 1d, convert to 2-column probability-like scores
        if dec.ndim == 1:
            dec = np.vstack([-dec, dec]).T
        # to map to "scores" for each class, return as-is
        return dec
    else:
        raise RuntimeError("Estimator has neither predict_proba nor decision_function.")

# Get scores for each model
y_score_rf = get_proba_or_decision(rf, X_test_proc)        # shape (n_samples, n_classes)
y_score_xgb = get_proba_or_decision(xgb_model, X_test_proc)

# Compute and print multiclass AUCs
print("\nRandom Forest ROC AUC (macro / micro):",
      roc_auc_score(y_test_bin, y_score_rf, average="macro"),
      "/", roc_auc_score(y_test_bin, y_score_rf, average="micro"))

print("XGBoost ROC AUC (macro / micro):",
      roc_auc_score(y_test_bin, y_score_xgb, average="macro"),
      "/", roc_auc_score(y_test_bin, y_score_xgb, average="micro"))

# Plotting function: per-class curves + micro-average
def plot_multiclass_roc(y_test_bin, y_score, title, ax):
    fpr = dict(); tpr = dict(); roc_auc = dict()
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    # micro-average
    fpr["micro"], tpr["micro"], _ = roc_curve(y_test_bin.ravel(), y_score.ravel())
    roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

    # Plot
    ax.plot(fpr["micro"], tpr["micro"],
            label=f"micro-average (AUC = {roc_auc['micro']:.3f})",
            linestyle=':', linewidth=2)
    for i in range(n_classes):
        ax.plot(fpr[i], tpr[i], lw=1.5, label=f"Class {classes[i]} (AUC = {roc_auc[i]:.3f})")

    ax.plot([0,1], [0,1], 'k--', lw=1)
    ax.set_xlim([-0.01, 1.01])
    ax.set_ylim([-0.01, 1.01])
    ax.set_xlabel("False Positive Rate")
    ax.set_ylabel("True Positive Rate")
    ax.set_title(title)
    ax.legend(loc="lower right", fontsize="small")

# Make side-by-side plots
fig, axes = plt.subplots(1, 2, figsize=(14,6))
plot_multiclass_roc(y_test_bin, y_score_rf, "Random Forest — One-vs-Rest ROC", axes[0])
plot_multiclass_roc(y_test_bin, y_score_xgb, "XGBoost — One-vs-Rest ROC", axes[1])
plt.tight_layout()
plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix, precision_recall_curve, average_precision_score
from sklearn.preprocessing import label_binarize
import itertools
import seaborn as sns

# Prepare
classes = np.unique(y_test)
n_classes = len(classes)
class_names = [str(c) for c in classes]  # readable labels

# 1) Target class distribution (plot from original y or y_test)
plt.figure(figsize=(6,4))
pd.Series(y).map({i: cls for i, cls in enumerate(le.classes_) if i < len(le.classes_)}).value_counts().reindex(le.classes_).plot(kind='bar')
plt.title("Readmitted - Class Distribution (full dataset)")
plt.xlabel("Readmitted")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# 2) Normalized confusion matrix for Random Forest
y_pred_rf = rf.predict(X_test_proc)
cm = confusion_matrix(y_test, y_pred_rf)        # raw counts
cm_norm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]  # normalize by true class (rows)

plt.figure(figsize=(7,5))
sns.heatmap(cm_norm, annot=True, fmt=".2f", cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.title('Random Forest - Normalized Confusion Matrix')
plt.tight_layout()
plt.show()

# 3) Precision-Recall curves (one-vs-rest) for Random Forest
# Binarize true labels
y_test_bin = label_binarize(y_test, classes=classes)

# Get probability scores (or decision scores) for RF
if hasattr(rf, "predict_proba"):
    y_score = rf.predict_proba(X_test_proc)
else:
    y_score = rf.decision_function(X_test_proc)
    if y_score.ndim == 1:
        y_score = np.vstack([-y_score, y_score]).T

# Compute PR curve and average precision for each class
precision = dict()
recall = dict()
avg_prec = dict()
for i in range(n_classes):
    precision[i], recall[i], _ = precision_recall_curve(y_test_bin[:, i], y_score[:, i])
    avg_prec[i] = average_precision_score(y_test_bin[:, i], y_score[:, i])

# micro-average
precision["micro"], recall["micro"], _ = precision_recall_curve(y_test_bin.ravel(), y_score.ravel())
avg_prec["micro"] = average_precision_score(y_test_bin, y_score, average="micro")

# Plot PR curves
plt.figure(figsize=(8,6))
plt.plot(recall["micro"], precision["micro"], linestyle=':', label=f"micro-average (AP = {avg_prec['micro']:.3f})", linewidth=2)
for i in range(n_classes):
    plt.plot(recall[i], precision[i], lw=1.5, label=f"Class {class_names[i]} (AP = {avg_prec[i]:.3f})")

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Random Forest - Precision-Recall curves (One-vs-Rest)')
plt.legend(loc='lower left', fontsize='small')
plt.xlim([-0.01,1.01])
plt.ylim([-0.01,1.01])
plt.tight_layout()
plt.show()
```
